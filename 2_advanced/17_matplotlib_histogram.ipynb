```python
#Importing Library
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_INA = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == "Indonesia")].copy()
df_DA_MAS = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == "Malaysia")].copy()
df_DA_USA = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == "United States")].copy()
df_DA_IRE = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == "Ireland")].copy()
df_DA_TUR = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == "Turkey")].copy()
df_DA_PAK = df[(df['job_title_short'] == "Data Analyst") & (df['job_country'] == "Pakistan")].copy()
```

```python
import numpy as np

# Mencari jumlah bin yang optimal
n = len(df_DA_USA['salary_year_avg'].dropna())
bins = int(np.ceil(np.log2(n) + 1))
```

```python
df_DA_USA['salary_year_avg'].plot(kind='hist', bins=40, edgecolor='black')
plt.xlim(0, 250000)

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.title('Distribution of USA Data Analyst Yearly Salary')
plt.ylabel('Number of Jobs')
plt.xlabel('Salaries')
```
