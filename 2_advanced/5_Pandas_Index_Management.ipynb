```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(10)
```

```python
df.index
```

```python
# mengubah nama index
df.index.name = 'job_index'
```

## Three Points to Cover

1. df.reset_index()
2. df.set_index()
3. df.sort_index()

```python
df_indo = df[df['job_country'] == 'Indonesia']
```

```python
df_indo.reset_index(inplace=True)
```

```python
df_indo.set_index('job_index', inplace=True)
```

```python
df_indo
```

```python
median_indo = df_indo.pivot_table(values='salary_year_avg', index='job_title_short', 
                                  aggfunc=['median', 'mean', 'min', 'max'])
median_indo
```

```python
median_indo.sort_values(by=('mean', 'salary_year_avg'), ascending=False)
```
